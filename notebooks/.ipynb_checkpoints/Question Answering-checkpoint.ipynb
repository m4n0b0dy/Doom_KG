```python
import spacy
from triples import *
nlp = spacy.load("en_core_web_lg")
```

```python
texts = ["How Biggie related to Afroman.",
    "Is Czarface related to DaBaby.",
    "Did Kanye perform with Eminem.",
    "Has Beyonce worked with Drake.",
    "How many songs does Tupac have with Kobe."]

triples = []
for text in texts:
    triples.append(processSentence(text, nlp))

printGraph(triples)
```

```python
rel_vecs = []
for triple in triples:
    subj, rel, obj = triple
    mtch, score = findClosestMatch(rel, nlp)
    rel_vecs.append((subj, mtch, obj, score, rel))
```
