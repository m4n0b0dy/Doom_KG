```python
import spacy
from neo4j import GraphDatabase
from triples import *
nlp = spacy.load("en_core_web_lg")
```

```python
url = "bolt://192.168.86.21:7687"
driver = GraphDatabase.driver(url, auth=("neo4j", "test"))
```

```python
texts = ["How Drake related to Eminem.",
    "What songs did Akon sing.",
    "Did Dababy perform with Afroman.",
    "Did Drake sing on Thursday",
    "How many songs does Beyonce have with Eminem."]

triples = []
for text in texts:
    triples.append(processSentence(text, nlp))
printGraph(triples)
```

```python
rel_vecs = []
for triple in triples:
    subj, rel, obj = triple
    mtch, score = findClosestMatch(rel, nlp, obj)
    rel_vecs.append((subj, mtch, obj))
    print(subj,mtch,obj,score)
```

```python
with driver.session() as session:
    for org_quest, rel_vec in zip(texts, rel_vecs):
        query, res = run_cypher_query(*rel_vec, session)
        print(org_quest)
        print(query.strip())
        for r in res.data():
            print(r)
        print('\n')
```
